## K-Means

### Pregunta 2

Ahora compararemos Dos algoritmos de clustering con los datos. Partiremos con K-means


Cuando usamos K-Means debemos definir previamente el número de clusters que queremos generar. Teniendo en cuenta estos datos, implemente el método del codo. ¿Cuántos clusters propone usar para este dataset? Escoja dos opciones y justifique su elección.

```python
from sklearn.cluster import KMeans

# Escriba su código aquí
# método del codo 

inercia = []

for i in range(1, 15):
    kmeans = KMeans(n_clusters=i, random_state=0)
    kmeans.fit(X)
    inercia.append(kmeans.inertia_)

plt.plot(range(1, 15), inercia)
plt.grid()
plt.title("Método del codo de 1 a 15 clusters")
plt.xlabel("Número de clusters")
plt.ylabel("Inercia")
plt.show()
```

> *Respuesta*

En función del gráfico obtenido del método del codo podemos ver que el codo se encuentra en 3 clusterns, también se podría considerar 4 clusters. Esto considerando que al pasar de 4 a 5 clusters no se observa un cambio significativo en la inercia.

Por lo tanto se proponen 3 y 4 clusters.

### Pregunta 3

 Genere un gráfico para cada uno de los dos `k` elegidos para tener una representación visual de los clusters. (usando los datos reducidos)

```python
k_escogido = 3
kmeans = KMeans(n_clusters=k_escogido, random_state=0)
kmeans.fit(X)
y_kmeans_3 = kmeans.predict(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans_3, cmap='viridis')
plt.title("Visualización PCA con KMeans")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.show()
```

```python
k_escogido = 4
kmeans = KMeans(n_clusters=k_escogido, random_state=0)
kmeans.fit(X)
y_kmeans_4 = kmeans.predict(X)

pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y_kmeans_4, cmap='viridis')
plt.title("Visualización PCA con KMeans")
plt.xlabel("Componente principal 1")
plt.ylabel("Componente principal 2")
plt.show()
```

## Clustering Jerárquico

### Pregunta 4

 Usando los datos anteriores, ejecute cada uno de los 4 métodos de clustering jerárquico: complete, single, average y ward, utilizando la metrica de distancia por defecto y visualice los dendrogramas formados. Adjunte el código necesario. No necesita generar los 4 dendrogramas en el mismo gráfico, pueden ser gráficos separados (individuales).

```python
from scipy.cluster.hierarchy import dendrogram, linkage

complete = linkage(X, method="complete")
single = linkage(X, method="single")
average = linkage(X, method="average")
ward = linkage(X, method="ward")
```

```python
# Escriba su código aquí

plt.figure(figsize=(15, 10))
plt.subplot(2, 2, 1)
dendrogram(complete)
plt.title("Complete")
plt.subplot(2, 2, 2)
dendrogram(single)
plt.title("Single")
plt.subplot(2, 2, 3)
dendrogram(average)
plt.title("Average")
plt.subplot(2, 2, 4)
dendrogram(ward)
plt.title("Ward")
plt.show()
```
